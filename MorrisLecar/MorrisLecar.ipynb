# Morris-Lecar Model




The Morris-Lecar model can be described by a system of two ordinary differential equations (ODEs) describing the changes of excitation of $Ca^{+2}$ (voltage-dependent) and $K^{+}$ (voltage-sensitive). The equations are the following:

<br>
$$
C\frac{dV}{dt} = I - g_{Ca} M_{ss} (V - V_{Ca}) - g_{K} W (V - V_{K}) +  - g_{L} (V - V_{L}) \\
\frac{dW}{dt} = \frac{W_{ss}(V) - W}{\tau_{W}(V)}
$$
<br>

where:

<br>
$$
M_{ss}(V) = \frac{1}{2} \bigg[ 1 + \tanh \bigg(\frac{V-V_{1}}{V_{2}} \bigg) \bigg] \\
W_{ss}(V) = \frac{1}{2} \bigg[ 1 + \tanh \bigg(\frac{V-V_{3}}{V_{4}} \bigg) \bigg]\\
\tau_{W}(V) = \frac{1}{\phi \cosh \bigg ( \frac{V-V_{3}}{2*V_{4}} \bigg)}
$$
<br>

Different parameters lead to different bifurcations in the model. Go the references to check with the parameters.

```python
# Hopf Bifurcation

# Import the module with the Hodgkin-Huxley model
from morris_lecar import MorrisLecar

# Set the parameters and create the model
params = {
    "phi":0.04,
    "V1":-1.2,
    "V2":18,
    "V3":2,
    "V4":30,
    "I":110,
    "time":1000,
}
model = MorrisLecar(**params)

# Run the model
model.run([-65, 0])

# Plot the results
model.plot()

# Plot the phase plane
model.phase_plane()

# Print the model parameters
# print(model)
```
