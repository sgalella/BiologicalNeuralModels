# Chialvo Map

The Chialvo Map is a two-dimensional map following the expression:

<br>
$$
x_{n+1} = x_n^2 \exp{(y_n - x_n)} + k \\
y_{n+1} = ay_n - bx_n + c
$$
<br>

where $x$ denotes the membrane potential of the neuron, $y$ is the recovery variable, $k$ a time-dependent additive perturbation, $a$ the time constant of recovery, $b$ the activation-dependence of the recovery process and $c$, the offset of the recovery. Parameters $a$ and $b$ should be smaller than 1.

The nullclines of the system are given when $x_{n+1} = x_{n}$ ($x^*$) and $y_{n+1} = y_{n}$ ($y^*$): 

<br>
$$
y^* = x^* + \ln{(x^* - k)} - 2\ln{(x^*)}\\
y^* = \frac{c - bx^*}{(1 - a)}
$$
<br>

```python
# Add temporary path to code
import sys
sys.path.append("../")

# Import the module with the Rulkov Map model
from neural_models import ChialvoMap

# Create the model
model = ChialvoMap(a=0.89, b=0.18, c=0.28, k=0.03)

# Print the model parameters
print(model)

# Run the model
model.run(X0=[0, 0], t=600)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Plot the results
f, ax1 = plt.subplots(figsize=(12, 5))
ax1.plot(model.tvec, model.x, color="royalblue")
ax1.set_xlabel('time [t]', fontsize=12)
ax1.set_ylabel(r"$x_n$", fontsize=15)
ax1.tick_params(axis='y', labelcolor='royalblue')
plt.grid(alpha=0.3)
ax2 = ax1.twinx()
ax2.plot(model.tvec, model.y, color="goldenrod")
ax2.set_ylabel(r"$y_n$", fontsize=15)
ax2.tick_params(axis='y', labelcolor='goldenrod')
plt.show()
```

```python
# Plot the phase space
plt.figure(figsize=(6, 4))
plt.plot(model.x, model.y, color='cornflowerblue')
plt.xlabel(r"$x_n$", fontsize=15)
plt.ylabel(r"$y_n$", fontsize=15)
plt.grid(alpha=0.3)
plt.show()
```
