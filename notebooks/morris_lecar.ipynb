# Morris-Lecar Model




The Morris-Lecar model can be described by a system of two ordinary differential equations (ODEs) describing the changes of excitation of $Ca^{+2}$ (voltage-dependent) and $K^{+}$ (voltage-sensitive). The equations are the following:

<br>
$$
C\frac{dV}{dt} = I - g_{Ca} M_{ss} (V - V_{Ca}) - g_{K} W (V - V_{K}) +  - g_{L} (V - V_{L}) \\
\frac{dW}{dt} = \frac{W_{ss}(V) - W}{\tau_{W}(V)}
$$
<br>

where:

<br>
$$
M_{ss}(V) = \frac{1}{2} \bigg[ 1 + \tanh \bigg(\frac{V-V_{1}}{V_{2}} \bigg) \bigg] \\
W_{ss}(V) = \frac{1}{2} \bigg[ 1 + \tanh \bigg(\frac{V-V_{3}}{V_{4}} \bigg) \bigg]\\
\tau_{W}(V) = \frac{1}{\phi \cosh \bigg ( \frac{V-V_{3}}{2*V_{4}} \bigg)}
$$
<br>

Different parameters lead to different bifurcations in the model. Go the references to check with the parameters.

```python
# Add temporary path to code
import sys
sys.path.append("../")

### Hopf Bifurcation

# Import the module with the Hodgkin-Huxley model
from neural_models import MorrisLecar

# Create the model
model = MorrisLecar(phi=0.04, V1=-1.2, V2=12, V3=2, V4=30)

# Run the model
model.run(X0=[-65, 0], current=110, t=1000)

# Print the model parameters
# print(model)
```

```python
import matplotlib.pyplot as plt

# Plot the results
f, ([ax1, ax2]) = plt.subplots(1, 2, figsize=(13, 3))

# Plot V(t)
ax1.plot(model.tvec, model.V, color='royalblue')
ax1.set_title("V", fontsize=15)
ax1.set_ylabel("Voltage [mV]", fontsize=12)
ax1.grid(alpha=0.3)

# Plot N(t)
ax2.plot(model.tvec, model.N, color='lightcoral')
ax2.set_title("N", fontsize=15)
ax2.set_ylabel("N(t)", fontsize=12)
ax2.grid(alpha=0.3)
plt.show()
```

```python
# Plot the phase plane
plt.figure(figsize=(7, 4))
    
# Plot N over Voltage
plt.plot(model.V, model.N, color='royalblue')
plt.xlabel("Voltage [mV]", fontsize=12)
plt.ylabel("N", fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
