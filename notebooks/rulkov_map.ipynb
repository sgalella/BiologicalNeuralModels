# Rulkov Map

The Rulkov Map is a low-dimensional system of discrete maps having the following expression:

<br>
$$
x_{n+1} = f(x_n, y_n) \\
y_{n+1} = y_n - \mu (x_n + 1) + \mu \sigma_n
$$
<br>

where $x$ denotes the membrane potential of the neuron, $y$ is a slow dynamical variable and $\sigma$ is a parameter expressing external influences (e.g. current).. $\mu$ controls the time evolution of $y$.  The function $f(x_n, y_n)$ is a piecewise function described as:

<br>
$$
f(x_n, y_n) = 
\begin{cases}
\frac{\alpha}{(1-x)} + y \ \ &\text{if } x \leq 0 \\
\alpha + y \ \ &\text{if } 0 < x < \alpha + y \\
-1  \ \ &\text{if } x \geq \alpha + y \\
\end{cases}
$$
<br>

We can study the stability of the fixed points for $x \leq 0$. The convergence can be defined as $x_{n+1} \approx x_n$, obtaining the following expression for the fixed points:


<br>
$$
x_{*} = \frac{(1-y) \pm \sqrt{y^2 -2y -(4\alpha  -1)}}{2}
$$
<br>

Evaluating the fixed points in the first derivative of $f(x_n, y_n)$ for $x \leq 0$, we have a stable fixed point if $|f(x_{*})|$ < 1, and a unstable fixed point if $|f(x_{*})|$ > 1.


```python
# Add temporary path to code
import sys
sys.path.append("../")

# Import the module with the Rulkov Map model
from neural_models import RulkovMap

# Set the parameters and create the model
params = {
    "alpha":5,
    "mu": 0.5,
    "sigma": 0.5
}

model = RulkovMap(**params)

# Run the model
model.run(t=100)
```

```python
import matplotlib.pyplot as plt

# Plot the results
f, ax1 = plt.subplots(figsize=(12, 5))
ax1.plot(model.tvec, model.x, color="royalblue")
ax1.set_xlabel('time [t]', fontsize=12)
ax1.set_ylabel(r"$x_n$", fontsize=15)
ax1.tick_params(axis='y', labelcolor='royalblue')
plt.grid(alpha=0.3)
ax2 = ax1.twinx()
ax2.plot(model.tvec, model.y, color="goldenrod")
ax2.set_ylabel(r"$y_n$", fontsize=15)
ax2.tick_params(axis='y', labelcolor='goldenrod')
plt.show()
```
